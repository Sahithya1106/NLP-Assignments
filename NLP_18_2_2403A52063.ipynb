```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization")
```

```python
text = """Artificial Intelligence is transforming the world by enabling machines to perform tasks that usually require human intelligence such as learning, reasoning, and decision making."""

summary = summarizer(text, max_length=50, min_length=20, do_sample=False)

print(summary[0]['summary_text'])
```

```python
!pip install transformers==4.40.2
!pip install torch --quiet
!pip install sentencepiece --quiet

from transformers import pipeline

summarizer = pipeline("summarization", device=-1)

text = """Artificial Intelligence is transforming industries by enabling machines to perform tasks that typically require human intelligence. It is widely used in healthcare, finance, and education."""

summary = summarizer(text[:500], max_length=50, min_length=20, do_sample=False)

print(summary[0]['summary_text'])
```

```python
text1 = """Paste a news article (150–300 words)"""
text2 = """Paste a technical paragraph"""
text3 = """Paste a story"""
text4 = """Paste a conversation"""
text5 = """Paste a blog or general content"""
```

```python
texts = [text1, text2, text3, text4, text5]
```

```python
for i, txt in enumerate(texts):
    print(f"\n====== TEXT {i+1} ======\n")
    print(txt)

    summary = summarizer(txt[:800], max_length=100, min_length=30, do_sample=False)

    print(f"\n------ SUMMARY {i+1} ------\n")
    print(summary[0]['summary_text'])
```

```python
text = text1

print("\nSHORT SUMMARY:\n")
print(summarizer(text[:800], max_length=50, min_length=20, do_sample=False)[0]['summary_text'])

print("\nMEDIUM SUMMARY:\n")
print(summarizer(text[:800], max_length=100, min_length=40, do_sample=False)[0]['summary_text'])

print("\nLONG SUMMARY:\n")
print(summarizer(text[:800], max_length=150, min_length=60, do_sample=False)[0]['summary_text'])
```

```python
long_text = """Paste a very long paragraph (400+ words)"""

summary = summarizer(long_text[:1000], max_length=150, min_length=50, do_sample=False)

print(summary[0]['summary_text'])
```
